```python
%sql --module letslookatthedata
SELECT * FROM [go-de-internal:democases_go_reply_internal.churn_4_created_data]
```

```python
churn = bq.Query(letslookatthedata).to_dataframe()
churn.describe()
```

```python
churn.sort_values(by=['timestamp'])
```

```python
shuffled = churn.sample(frac=1)
trainsize = int(len(shuffled['timestamp']) * 0.70)
validsize = int(len(shuffled['timestamp']) * 0.15)

df_train = shuffled.iloc[:trainsize, :]
df_valid = shuffled.iloc[trainsize:(trainsize+validsize), :]
df_test = shuffled.iloc[(trainsize+validsize):, :]
```

```python
def to_csv(df, filename):
  outdf = df.copy(deep=False)
  
  # reorder columns so that target is first column
  cols = outdf.columns.tolist()
  cols.remove('timestamp')
  cols.insert(0, 'timestamp')
  print (cols)  # new order of columns
  outdf = outdf[cols]
  outdf.to_csv(filename, header=True, index_label=False, index=False)
  
  print outdf.shape

to_csv(churn, 'churn-timedepenent-all.csv')
to_csv(df_train, 'churn-timedepenent-train.csv')
to_csv(df_valid, 'churn-timedepenent-valid.csv')
to_csv(df_test, 'churn-timedepenent-test.csv')
```

```python
from pandas import read_csv
from datetime import datetime
# load data
dataset = read_csv('churn-timedepenent-all.csv', index_col=False)
#dataset.drop('user_id')
dataset = dataset[['timestamp', 'user_id', 'event_id', 'event_name', 'churn']]
dataset.head(5)
```

```python
# manually specify column names
dataset.columns = ['event_id', 'event_name', 'churn']
dataset.index.name = 'timestamp'
# mark all NA values with 0
dataset['churn_prediction'].fillna(0, inplace=True)
# summarize first 5 rows
print(dataset.head(5))
# save to file
dataset.to_csv('churn-for-LSTM.csv')
```

```python
from pandas import read_csv
from matplotlib import pyplot
# load dataset
dataset = read_csv('pollution.csv', header=0, index_col=0)
values = dataset.values
# specify columns to plot
groups = [0, 1, 2, 3, 5, 6, 7]
i = 1
# plot each column
pyplot.figure()
for group in groups:
    pyplot.subplot(len(groups), 1, i)
    pyplot.plot(values[:, group])
    pyplot.title(dataset.columns[group], y=0.5, loc='right')
    i += 1
pyplot.show()
```
